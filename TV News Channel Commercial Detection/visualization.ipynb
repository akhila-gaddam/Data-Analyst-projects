## linear dimensionality reductions - PCA 

```python
import matplotlib.pyplot as plt
import numpy as np

X, y = df.iloc[:,2:].to_numpy(), df.target.to_numpy()

# y = y.astype('str')
# y[y == '1'] = 'r'
# y[y == '0'] = 'g'
```

```python
import numpy as np
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
pca.fit(X)

linear_reduced_data = pca.transform(X)

plt.scatter(linear_reduced_data.take((0), axis=1), 
            linear_reduced_data.take((1), axis=1), 
            c = y, 
            alpha=0.3)
plt.show()
```

## Non linear dimensionality reductions - KPCA 

```python
from sklearn.datasets import load_digits
from sklearn.decomposition import KernelPCA

kpca = KernelPCA(n_components = 2,
                        kernel='poly',)
                        # kernel='rbf')
                        # kernel = 'sigmoid',)
                        # kernel = 'cosine',)
kpca.fit(X)
non_linear_reduced_data = kpca.transform(X)

plt.scatter(non_linear_reduced_data.take((0), axis=1), 
            non_linear_reduced_data.take((1), axis=1), 
            c = y, 
            alpha=0.3)
plt.show()
```
